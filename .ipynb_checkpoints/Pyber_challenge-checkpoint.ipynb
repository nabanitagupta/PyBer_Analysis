```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as sts
import matplotlib as mpl
```

```python
# Files to load
city_data_to_load = "Resources/city_data.csv"
ride_data_to_load = "Resources/ride_data.csv"
```

```python
# Read the city data file and store it in a pandas DataFrame.
city_data_df = pd.read_csv(city_data_to_load)
city_data_df.head(10)
```

```python
# Read the ride data file and store it in a pandas DataFrame.
ride_data_df = pd.read_csv(ride_data_to_load)
ride_data_df.count()
```

```python
ride_data_df.head()
```

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the DataFrame
pyber_data_df.head()
```

```python
pyber_data_df.count()
```

```python
# Get the number of rides for each cities.
total_rides = pyber_data_df.groupby(["type"]).count()["ride_id"]
total_rides
```

```python
# Get the total number of drivers for each cities. Notice city data had to be used
total_drivers = city_data_df.groupby('type')['driver_count'].sum()
total_drivers
```

```python
total_fares = pyber_data_df.groupby('type')['fare'].sum()
total_fares
```

```python
total_fares.sum()
```

```python
# Create the Urban, ruraln and suburban city DataFrame.
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
```

```python
rural_cities_df
```

```python
# Calculate Average Fare per Ride for each city type.
average_fare_per_ride = pyber_data_df.groupby('type')['fare'].mean()
average_fare_per_ride
```

```python
# Calculate Average Fare per driver for each city type.
average_fare_per_driver = total_fares/total_drivers
# Naming the series index and values Average Fare per driver for each city type.
average_fare_per_driver = pd.DataFrame({"type": average_fare_per_driver.index, "Average fare per driver": average_fare_per_driver.values})
average_fare_per_driver
```

```python
# Building the summary dataframe
summary_df = pd.merge(total_rides,total_drivers,left_on="type", right_on="type")
summary_df = pd.merge(summary_df, total_fares, left_on="type", right_on="type")
summary_df = pd.merge(summary_df, average_fare_per_ride, left_on="type", right_on="type")
summary_df = pd.merge(summary_df, average_fare_per_driver, left_on="type", right_on="type")
summary_df
```

```python
# Renaming columns in summary dataframe
summary_df.rename(columns={'ride_id':'Total Rides', 'driver_count':'Total Drivers','fare_x': 'Total Fares', 'fare_y': 'Average Fare per Ride'}, inplace=True)
summary_df
```

```python
# Setting the index 
summary_df = summary_df.set_index('type')
summary_df
```

```python
# Delete the index name
del summary_df.index.name
summary_df
```

```python
# Formatting/styling specific columns
summary_df['Total Fares'] = summary_df['Total Fares'].apply('${:,.2f}'.format)
summary_df['Average Fare per Ride'] = summary_df['Average Fare per Ride'].apply('${:.2f}'.format)
summary_df['Average fare per driver'] = summary_df['Average fare per driver'].apply('${:.2f}'.format)
summary_df
```

```python
# Formatting/styling specific columns
summary_df['Total Rides'] = summary_df['Total Rides'].apply('{:,}'.format)
summary_df['Total Drivers'] = summary_df['Total Drivers'].apply('{:,}'.format)
summary_df
```

```python
pyber_data_df
```

```python
# Renaming columns in merged dataframe
pyber_data_df.rename(columns={'city': 'City', 'date':'Date','fare':'Fare', 'ride_id': 'Ride Id','driver_count': 'No. Drivers', 'type':'City Type'}, inplace=True)
pyber_data_df
```

```python
# Set the index to the Date column
pyber_data_df = pyber_data_df.set_index('Date')
pyber_data_df
```

```python
# Delete some columns
fare1_df = pyber_data_df.drop(columns=['Ride Id', 'No. Drivers', 'City'])
fare1_df
```

```python
# Copy some columns to a new df. Default is deep copy. 
fare_df = pyber_data_df.copy()
fare_df
```

```python
# Delete some columns
fare_df = fare_df.drop(columns=['Ride Id', 'No. Drivers', 'City'])
fare_df
```

```python
# Set the index to the datetime data type
fare_df.index = pd.to_datetime(fare_df.index)
fare_df
```

```python
# Check the DataFrame using the info() 
# method to make sure the index is a datetime data type
print("Summary of the basic information about this DataFrame and its data:")
print(fare_df.info())
```

```python
# Calculate the sum() of fares by the type of city and date 
# using groupby() to create a new DataFrame
fare_city_type = fare_df.groupby(['City Type', 'Date'])
fare_city_type.first()
```

```python
# Calculate the sum() of fares by the type of city and date 
# using groupby() to create a new DataFrame
total_fare_city_type = fare_city_type ['Fare'].sum ()
total_fare_city_type.tail()
```

```python
# Reset index
total_fare_city_type = total_fare_city_type.reset_index() # When we reset the index, the old index is added as a column, and a new sequential index is used
```

```python
total_fare_city_type
```

```python
# Create a pivot table DataFrame
table = pd.pivot_table(total_fare_city_type, index =['Date'], columns =['City Type'], values ='Fare') 
table.tail()
```

```python
# Create a new DataFrame from the pivot table DataFrame on the given dates
table_new = table.loc['2019-01-01':'2019-04-28']
table_new.head()
```

```python
print(table_new.index)
```

```python
# Converting pivot table to pandas dataframe
table_new.columns.name = None    #remove categories
table_new2 = pd.DataFrame (table_new)
```

```python
table_new2.head()
```

```python
# New dataframe to sum fare for each week
table_new3 =table_new2.resample('7D').sum()
```

```python
table_new3
```

```python
# Plotting the data
table_new4 = table_new3.reset_index ()
plt.style.use('fivethirtyeight')
fig, ax = plt.subplots(figsize=(15, 6))
ax = plt.gca()
table_new4.plot(kind = 'line', x='Date', y='Rural', ax = ax)
table_new4.plot(kind = 'line', x='Date', y='Suburban', ax = ax)
table_new4.plot(kind = 'line', x='Date', y='Urban', ax = ax)
ax.set_title('Total Fare by City Type')
ax.set_ylabel('Fare ($USD)')
ax.set_xlabel('Month')
ax.grid
plt.show()
# Save the figure.
plt.savefig("analysis/Fig6.png")
```
